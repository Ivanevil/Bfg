```python
import pandas as pd # 5 zad
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
myfile = pd.read_csv("data_var02.txt")
myfile.head()

fid = open("data_var02.txt", "r")
lines = fid.readlines()
fid.close()
y = [float(lines[i].split(',')[1]) for i in range(1, len(lines))]
x = [i for i in range(1, len(y)+1)]
n=len(y) #    
s=sum(y) #   y
s1=sum([1/y[i] for i in  range(0,n)]) #   1/x
s2=sum([(1/y[i])**2 for i in  range(0,n)]) #   (1/x)**2
s3=sum([y[i]/y[i]  for i in range(0,n)])  #  y/x                   
a= round((s*s2-s1*s3)/(n*s2-s1**2),3) #      
b=round((n*s3-s1*s)/(n*s2-s1**2),3)#  b    
s4=[a+b/y[i] for i in range(0,n)] #                  



plt.plot(x, y)
plt.plot(x, s4)


plt.grid(True)
plt.show()
```

```python
import pylab
from matplotlib import mlab
fid=open('C:/Users/shum-vokrug/Downloads/data_var25.txt','r')
lines=fid.readlines()
fid.close()
n=5
y=[0 for i in range(n)]
for i in range (n):
    y[i]=float(lines[i+1].split(',')[1])
p=[0 for i in range(n)]
p[0]=y[0]
l=n
for j in range(1,n):
    for k in range(l-1):
        y[k]=(y[k+1]-y[k])/j
    p[j]=y[0]
    l=l-1
def fun(x):
    z=p[0]
    a=1
    for m in range(n-1):
        a=a*(x-m)
        z=z+p[m+1]*a
    return z
xlist=mlab.frange(-0.5,4.5,0.1)
ylist=[fun(x) for x in xlist]
pylab.plot(xlist,ylist)
for r in range(n):
    pylab.plot(r,float(lines[r+1].split(',')[1]),'.r')
pylab.grid()
pylab.show()
```
